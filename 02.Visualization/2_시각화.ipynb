pandas의 merge로 합치기 

방법 4가지 

1. inner join 
2. left outer join
3. right outer join
4. full outer join


```python
# inner join
pd.merge(vp, vq, on='Product', how='inner')
```

```python
# left outer join
pd.merge(vp, vq, on='Product', how='left')
```

```python
# right outer join
pd.merge(vp, vq, on='Product', how='right')
```

```python
# full outer join
pd.merge(vp, vq, on='Product', how='outer')
```
